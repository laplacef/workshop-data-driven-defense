```python
import sys
import pprint

sys.path.append("../")

from src.database import DBConn

db_dir = "../db/"
db_name = "edge_iiotset.db"

iiotset_db = DBConn(db_name, db_dir)
```

## What does the data look like?

We can start by looking at the tables in the database and getting a sense of the shape of the data. We can also look at the first few rows of each table to get a sense of the data and what it looks like. Let's break this down by attack and non-attack data.

### Attack Traffic

```python
attack_tables = {
    table: iiotset_db.get_shape(table)
    for table in iiotset_db.table_names()
    if table.endswith("attack")
}
pprint.pprint(attack_tables)
```

```python
pprint.pprint(iiotset_db.get_columns("backdoor_attack"))
```

```python
for t in attack_tables:
    print(t)
    display(iiotset_db.get_rows(t, 2))
    print("\n\n")
```

## Normal Traffic

```python
normal_tables = {
    table: iiotset_db.get_shape(table)
    for table in iiotset_db.table_names()
    if not (table.endswith("attack") or table.endswith("dataset"))
}
pprint.pprint(normal_tables)
```

```python
pprint.pprint(iiotset_db.get_columns("distance"))
```

```python
for t in normal_tables:
    print(t)
    display(iiotset_db.get_rows(t, 2))
    print("\n\n")
```
